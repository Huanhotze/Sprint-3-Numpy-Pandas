## Exercici 1
Crea una funció que donat un Array d’una dimensió, et faci un resum estadístic bàsic de les dades. Si detecta que l’array té més d’una dimensió, ha de mostrar un missatge d’error.

```python
def crea_array():
  dimensiones = random.randint(1, 2)
  values = random.randint(3, 10)
  return (np.random.random((dimensiones,values))*10).astype(int)

def info_array(a):
  #print(a.shape) #Array dimensions
  #print(len(a[0])) #Length of array 
  print(f"Nº de dimensiones: {a.ndim}") #Number of array dimensions 
  print(f"Nº de elementos: {a.size}") #Number of array elements
  print(f"Datatype de los elemetos: {a.dtype}") #Data type of array elements
  print(f"La suma de los elementos: {a.sum()}") #Array-wise sum
  print(f"El minimo: {a.min()}") #Array-wise minimum value 
  print(f"El maximo: {a.max()}") #Maximum value of an array row
  #print(f"La suma acumulada: {a.cumsum(axis=1)}") #Cumulative sum of the elements
  print(f"El promedio: {a.mean()}") #Mean
  print(f"La mediana: {np.median(a)}") #Median
  print(f"El coeficiente de correlación: {np.corrcoef(a)}") #Correlation coefficient
  print(f"La desviación estandard: {np.std(a)}") #Standard deviation


arr = crea_array()
print(arr)
if np.shape(arr)[0] == 2:
  print("ERROR!!!")
else:
  info_array(arr)
```

## Exercici 4
Implementa manualment una funció que calculi el coeficient de correlació. Informa-te'n sobre els seus usos i interpretació.


```python
# Variación interanual T1 PIB España  (2001-2021) (fte. INE)
pib = np.array([3.9,2.7,3,3.1,3.7,4.1,3.6,0.9,-3.8,0.2,-0.8,-3,-1.4,1.4,3.8,3,3,2.3,2.1,-10.8,5.1
])
# Tasa de paro anual (2001 - 2021) (fte. Expansion)
paro = np.array([10.94,11.55,11.99,11.5,10.17,9.03,8.42,9.6,17.24,19.84,21.08,24.19,26.94,25.93,23.78,21,18.75,16.74,14.7,14.41,15.98
])

#Número de valores o elementos Σxy
n = pib.size
#la suma de los productos de las puntuaciones emparejadas
producto = pib * paro
sum_producto = producto.sum()
#la suma de puntuaciones x
sum_pib = pib.sum()
#la suma de puntuaciones y
sum_paro = paro.sum()
#Suma de cuadrados Puntuación x
cuadrado_pib = pib ** 2
sum_2_pib = cuadrado_pib.sum()
#Suma de cuadrados Puntuación y
cuadrado_paro = paro ** 2
sum_2_paro = cuadrado_paro.sum()

#Calculo correlacion
correlacion = ((n*sum_producto)-(sum_pib*sum_paro)) / math.sqrt(((n*sum_2_pib)-(sum_pib**2))*((n*sum_2_paro)-(sum_paro**2)))
print(correlacion)

#Comprobación calculos manuales
print(np.corrcoef(pib, paro))
```

**Correlación:** La correlación trata de establecer la relación o dependencia que existe entre las dos variables que intervienen en una distribución bidimensional. Es decir, determinar si los cambios en una de las variables influyen en los cambios de la otra. En caso de que suceda, diremos que las variables están correlacionadas o que hay correlación entre ellas. La correlación es positiva cuando los valores de las variables aumenta juntos; y es negativa cuando un valor de una variable se reduce cuando el valor de la otra variable aumenta.
